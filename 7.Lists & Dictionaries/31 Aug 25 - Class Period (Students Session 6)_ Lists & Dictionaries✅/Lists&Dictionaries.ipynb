### Dictionary

```python
# Let say we want to create a dictionary of test scores for each of the students
score_dict = {'Bell':75,
            'John':96,
            'Shinta':83.34345,
            'Mark':55,
            'Kurnia':90
            }
print(score_dict)
```

```python
# Let say we want to create a dictionary of test scores for each of the students
score_dict_1 = {'Bell':[75,True],
            'John':[96,90],
            'Shinta':[83.34345,91],
            'Mark':[55,92],
            'Kurnia':[90,93]
            }
print(score_dict_1)
```

```python
# Let say we want to create a dictionary of test scores for each of the students
score_dict_2 = {'Bell':[75,True],
            'John':{
                'alamat':'JL. Sampoerna No 1 Indonesia',
                'age':25
            }
          }
print(score_dict_2)
```

```python
# We can access the score for each of the students via its key (name of the student)
print(score_dict['John'])
print(score_dict['Mark'])
```

```python
score_dict
```

```python
# We can add the element if there's new entry
score_dict['Bell'] = 77
print(score_dict)
```

```python
# We can also delete a specific entry
del score_dict['Shinta']
print(score_dict)
```

```python
# We'are also able to access the sequences of our dictionary keys and values
print(score_dict.keys())
print(score_dict.values())
```

```python
# change key name in dictionary
# dictionary[new_key] = dictionary[old_key]
# del dictionary[old_key]
score_dict['Haekal'] = score_dict.pop('John')
score_dict
```

```python
score_dict['Anwar'] = score_dict['Kurnia']
del score_dict['Kurnia']
```

```python
def parse_input(user_input):
    result = []
    for token in user_input.split():
        if token.isdigit():
            result.append(int(token))  # integer
        else:
            try:
                result.append(float(token))  # float
            except ValueError:
                if token.lower() == "true":
                    result.append(True)
                elif token.lower() == "false":
                    result.append(False)
                else:
                    result.append(token)  # string
    return result
```

```python
user_input = input("Masukkan data (contoh: 1 2 a True 4.5): ")
data_list = parse_input(user_input)

print("Hasil list:", data_list)
```
